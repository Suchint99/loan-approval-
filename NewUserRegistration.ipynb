```python
from tkinter import *
import mysql.connector

def nur():
    window1 = Tk()
    window1.title("User Registration Form")
    window1.geometry('500x300')
    un = Label(window1, text="Enter the User name",font=("Arial Bold", 10))
    pwd = Label(window1, text="Enter the Password",font=("Arial Bold", 10))
    name = Label(window1, text="Enter the Name",font=("Arial Bold", 10))
    address = Label(window1, text="Enter Address",font=("Arial Bold", 10))
    mobileno = Label(window1, text="Enter Mobile No",font=("Arial Bold", 10))
    email = Label(window1, text="Enter Email",font=("Arial Bold", 10))
    sq = Label(window1, text="Enter Security Question",font=("Arial Bold", 10))
    ans = Label(window1, text="Enter Answer",font=("Arial Bold", 10))
    txt1 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt2 = Entry(window1,width=30,show='*',font=("Arial Bold", 10))
    txt3 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt4 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt5 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt6 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt7 = Entry(window1,width=30,font=("Arial Bold", 10))
    txt8 = Entry(window1,width=30,font=("Arial Bold", 10))
    
        
    un.grid(column=1, row=0)
    txt1.grid(column=2, row=0)

    pwd.grid(column=1, row=1)
    txt2.grid(column=2, row=1)
    
    name.grid(column=1, row=2)
    txt3.grid(column=2, row=2)
    
    address.grid(column=1, row=3)
    txt4.grid(column=2, row=3)
    
    mobileno.grid(column=1, row=4)
    txt5.grid(column=2, row=4)
    
    email.grid(column=1, row=5)
    txt6.grid(column=2, row=5)
    
    sq.grid(column=1, row=6)
    txt7.grid(column=2, row=6)
    
    ans.grid(column=1, row=7)
    txt8.grid(column=2, row=7)
    def register():   
        un=txt1.get()
        pwd=txt2.get()
        name=txt3.get()
        address=txt4.get()
        mobileno=txt5.get()
        email=txt6.get()
        sq=txt7.get()
        ans=txt8.get()
        mycon1 = mysql.connector.connect(host = "localhost", user = "root", passwd = "", database="home_loan_and_house_price_prediction_db")
        cur = mycon1.cursor()
     
        sql ="insert into user_registration values(%s,%s,%s,%s,%s,%s,%s,%s)"
        val = (un,pwd,name,address,mobileno,email,sq,ans)
        cur.execute(sql,val)
        mycon1.commit()
        print ("one record added successfully")
    Register = Button(window1, text="Submit",font=("Arial Bold", 10),command=register)
    Register.grid(column=2, row=8)
   
    window1.mainloop()
nur()
```
