```python
import pandas as pd
from tkinter import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from tkinter import messagebox

def hpp():
    window4 = Tk()
    window4.title("House Price Prediction Form")
    window4.geometry('500x300')
    ha = Label(window4, text="Enter House Area ",font=("Arial Bold", 10))
    tf = Label(window4, text="Enter Total Floors",font=("Arial Bold", 10))
    
    aoh = Label(window4, text="Enter Age of House",font=("Arial Bold", 10))
    wea = Label(window4, text="Enter Water_Electricity_Availability",font=("Arial Bold", 10))
    paa = Label(window4, text="Enter Play_Area_Availability",font=("Arial Bold", 10))
    
    ta = Label(window4, text="Enter Terrace_Availability",font=("Arial Bold", 10))
    cpa = Label(window4, text="Enter Car_Parking_Availability",font=("Arial Bold", 10))
    sa = Label(window4, text="Enter Security_Availability",font=("Arial Bold", 10))
    
    pp = Label(window4, text="House Predicted Price is=",font=("Arial Bold", 10))
    
    txt1 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt2 = Entry(window4,width=30,show='*',font=("Arial Bold", 10))
    txt3 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt4 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt5 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt6 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt7 = Entry(window4,width=30,font=("Arial Bold", 10))
    txt8 = Entry(window4,width=30,font=("Arial Bold", 10))
    ha.grid(column=1, row=0)
    txt1.grid(column=2, row=0)

    tf.grid(column=1, row=1)
    txt2.grid(column=2, row=1)
    
    aoh.grid(column=1, row=2)
    txt3.grid(column=2, row=2)
    
    wea.grid(column=1, row=3)
    txt4.grid(column=2, row=3)
    
    paa.grid(column=1, row=4)
    txt5.grid(column=2, row=4)
    
    ta.grid(column=1, row=5)
    txt6.grid(column=2, row=5)
    
    cpa.grid(column=1, row=6)
    txt7.grid(column=2, row=6)
    
    sa.grid(column=1, row=7)
    txt8.grid(column=2, row=7)
    
    def predict():
        ha=int(txt1.get())
        tf=int(txt2.get())
        aoh=int(txt3.get())
        wea=int(txt4.get())
        paa=int(txt5.get())
        ta=int(txt6.get())
        cpa=int(txt7.get())
        sa=int(txt8.get())
        newx=[[ha,tf,aoh,wea,paa,ta,cpa,sa]]
        df = pd.read_csv("House_Price_Dataset.csv",sep = ",")
        # Now define x aand y
        y = df["Price"]
        x = df.drop(["Price"],axis=1)
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state=42)
        #Use Linear Regression algorithm 
        model = LinearRegression()
        model.fit(x_train,y_train)
        y_predict = model.predict(newx)
        #messagebox.showinfo('House Price Prediction', 'House Predicted Price is'+ y_predict)
        pp.config(text = y_predict)
        
    predict = Button(window4, text="Predict Price",font=("Arial Bold", 10),command=predict)
    predict.grid(column=2, row=8)
    pp.grid(column=2,row=9)
    window4.mainloop()
```
