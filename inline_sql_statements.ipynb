```python
from tkinter import *
from tkinter import messagebox
import mysql.connector
myconn = mysql.connector.connect(host = "localhost", user = "root",passwd = "",database="emsdb")  
cur = myconn.cursor()
def submit():
    un=txt1.get()
    pwd=txt2.get()
    name = txt3.get()
    city = txt4.get()
    qry="insert student_info values('"+un+"' ,'"+pwd+"','"+name+"','"+city+"')"
    try:  
    #inserting the values into the table  
        cur.execute(qry)
  
    #commit the transaction   
        myconn.commit()
        messagebox.showinfo('User Regestration', 'Record Inserted Successfuly')
      
    except:
        myconn.rollback()  
        
def delete_record():
    un=txt1.get()
   
    qry="delete from student_info where username ='"+un+"'"
    try:  
    #inserting the values into the table  
        cur.execute(qry)
  
    #commit the transaction   
        myconn.commit()
        messagebox.showinfo('Record Deletion','Record Deleted Successfuly')
      
    except:
        myconn.rollback() 

def search_record():
    un=txt1.get()
   
    qry="select * from student_info where username ='"+un+"'"
    cur.execute(qry)
    row = cur.fetchone()
    txt1.delete(0, END)
    txt1.insert(0,row[0])
    txt2.insert(0,row[1])
    txt3.insert(0,row[2])
    txt4.insert(0,row[3])
   
    
def cancel():
    txt1.delete(0, END)
    txt2.delete(0,END)
    txt1.focus()

window = Tk()
window.title("Student Regestration")
window.geometry('450x150')
lb1 = Label(window, text="Enter the User name",font=("Arial Bold", 10))
lb2 = Label(window, text="Enter the Password",font=("Arial Bold", 10))
lb3 = Label(window, text="Enter the name",font=("Arial Bold", 10))
lb4 = Label(window, text="Enter the city",font=("Arial Bold", 10))
txt1 = Entry(window,width=30,font=("Arial Bold", 10))
txt2 = Entry(window,width=30,font=("Arial Bold", 10))
txt3 = Entry(window,width=30,font=("Arial Bold", 10))
txt4 = Entry(window,width=30,font=("Arial Bold", 10))

submit = Button(window, text="Submit",font=("Arial Bold", 10),command=submit)
cancel = Button(window, text="Cancel",font=("Arial Bold", 10),command=cancel)
delete = Button(window, text="Delete",font=("Arial Bold", 10),command=delete_record)
search = Button(window, text="Search",font=("Arial Bold", 10),command=search_record)
lb1.grid(column=1, row=0)
txt1.grid(column=2, row=0)
lb2.grid(column=1, row=1)
txt2.grid(column=2, row=1)
lb3.grid(column=1, row=2)
txt3.grid(column=2, row=2)
lb4.grid(column=1, row=3)
txt4.grid(column=2, row=3)
submit.grid(column=1, row=4)
cancel.grid(column=2, row=4)
delete.grid(column=1, row=5)
search.grid(column=2, row=5)
window.mainloop()
```
